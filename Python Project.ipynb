   ## Analysis on the International Development Association (IDA) Credits and Grants
   
   
   ### The International Development Association (IDA) credits are public and publicly guaranteed debt extended by the World Bank Group. IDA provides development credits, grants and guarantees to its recipient member countries to help meet their development needs. Credits from IDA are at concessional rates. Data are in U.S. dollars calculated using historical rates. This dataset contains the latest available snapshot of the IDA Statement of Credits and Grants.

## Data description:

### Rows : 8,770   
### Columns : 30
### End of period : December 31st, 2019
### Period range : May 12, 1961 to December 31st, 2019

```python
ida_1 = ida.loc[:, ['Credit Number','Region', 'Country', 'Credit Status', 'Service Charge Rate', 'Currency of Commitment',
                   'Project ID', 'Original Principal Amount', 'Cancelled Amount', 'Undisbursed Amount', 'Disbursed Amount',
                   'Repaid to IDA', 'Due to IDA', 'Credits Held', 'First Repayment Date', 'Last Repayment Date', 
                    'Agreement Signing Date']]
```

```python
ida_1.columns
```

```python
type(ida['Agreement Signing Date'][0:])
```

```python
ida_1['agr_date'] = pd.to_datetime(ida['Agreement Signing Date'], format='%m/%d/%Y %H:%M')
ida_1['agr_date']
```

```python
ida_1['frst_date'] = pd.to_datetime(ida['First Repayment Date'], format='%m/%d/%Y %H:%M')
ida_1['frst_date']
```

```python
ida_1['Year'] = pd.DatetimeIndex(ida_1['agr_date']).year
ida_1.columns
```

```python
ida_1['Year1'] = pd.DatetimeIndex(ida_1['frst_date']).year
ida_1.columns
```

```python
loan_nb = ida_1.groupby(['Year'])['Credit Status'].count()
loan_amt = ida_1.groupby(['Year'])['Disbursed Amount'].sum()/1000000
loan_amt
```

# The evolution of loans approval over the years

```python
plt.figure(figsize=(15,8))
loan_nb.plot(label = 'Loan Numbers', color = 'r', legend=True)
loan_amt.plot(label = 'Loan Amount', secondary_y = True, color = 'b', legend=True)
plt.title('Evolution of approved loans over the years\n', fontsize = 15, color = 'b')
plt.xlabel('Years', fontsize = 15, color='b')
plt.xlim(1960,2020)
plt.show()
```

### Over the years we clearly see an increase of loans approved by IDA. Around 2015-2016 we see a considerable drop of the loans amount approved. 

## The number of loans approved by Region

### Top 3 regions having loans from IDA are : Africa, South Asia and East Asia & Pacific

#### We can see for East Asia and South Asia the line of total loan amount is on top of the number of loans, that means they have less loans but the loan amount is high. Let see on next the distribution of loans amount and loans number separately. 

```python
# Graph representing the proportionality between the number of loans and the loans amount distributed by each region

plt.figure(figsize=(15,7))
ida_1.groupby('Region')['Region'].count().sort_values(ascending = False).plot.bar(legend = True, label = 'Number of Loans', color='c', fontsize = 9)
ida_1.groupby('Region')['Credits Held'].sum().sort_values(ascending = False).plot(secondary_y=True, legend = True, label = 'Total Loan Amount', color='r', fontsize = 9)
plt.title('Number of loans per Region with total amount\n', fontsize=15, color = 'b')
plt.show()
```

## Loans number distribution per region

```python
# Loan number approved by region
plt.figure(figsize=(15,8))

pd.crosstab(ida_1.Year, ida_1.Region).plot()
plt.title('Loan number approved by region\n', fontsize=15, color='b')
plt.xlabel('Years')
plt.xlim(1960, 2020)
plt.ylabel('Number of loans')
plt.legend(loc='upper right', bbox_to_anchor=(1.65, 1), scatterpoints = 1, #markerfirst = False,
          shadow = True, facecolor = 'lightgray', title='Region', labelspacing=0.8)
plt.show()
```

```python
ida_1['Total_loan'] = ida_1['Undisbursed Amount'] + ida_1['Disbursed Amount']
ida_1
```

## The top 5 countries with more debt and with less debt from IDA

#### The 5 top countries with more debt are from region South Asia, East Asia and Africa
#### We see next that the top 5 countries in debt are coming from the regions previoulsy identified in distribution of loans amount and loans number

```python
top5 = ida_1.groupby('Country')['Credits Held'].sum().sort_values(ascending = False)[:5]
top5
```

```python
plt.figure(figsize=(15,8))
top5.plot.barh(color='y')
plt.title('Top 5 countries with more debt\n', fontsize = 15, color = 'b')
plt.xlabel('Countries', fontsize = 15, color = 'r')
plt.ylabel('Loans amount', fontsize = 15, color = 'r')
plt.legend()
plt.show()
```

## Top 5 countries with less debt from IDA

```python
top5_asc = ida_1.loc[ida_1['Credits Held']!=0, :]
top5asc = top5_asc.groupby('Country')['Credits Held'].sum().sort_values(ascending = True)[:5]
```

```python
top5asc
```

```python
plt.figure(figsize=(15,8))
top5asc.plot.bar(color='m')
plt.title('Top 5 countries with less debt\n', fontsize = 15, color = 'b')
plt.xlabel('Countries', fontsize = 15, color = 'r')
plt.ylabel('Loans amount', fontsize = 15, color = 'r')
plt.legend()
plt.show()
```
